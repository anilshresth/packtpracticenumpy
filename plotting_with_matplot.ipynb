```python
#lets create a polynmial graph
import numpy as np
import matplotlib.pyplot as plt
```

```python


#define a function
func=np.poly1d(np.array([1,2,3,4]).astype(float))

#define the x value to use
x=np.linspace(-10,10,30)

#use the x to get valuw from func
y=func(x)

#plot the values
plt.plot(x,y)

#define a title
plt.title('polyynomial graph',color='g')

#define the x labels
plt.xlabel('xaxis')

#define the y axis
plt.ylabel('yaxis')

#display the figure
plt.show()

```

```python
func=np.poly1d(np.array([1,2,3,4]).astype(float))

#first order derivative
func1=func.deriv(m=1)
x = np.linspace(-10, 10, 30)
y = func(x)
y1 = func1(x)
plt.plot(x,y,'ro',x,y1,'g^')#ro gives the red and circe to the points,similar with the -- that gives the --
plt.show()
```

```python
# creating the subplot,Create the first subplot of the polynomial with the subplot function. The first
# parameter of this function is the number of rows, the second parameter is the
# number of columns, and the third parameter is an index number starting with 1.
# Alternatively, you can combine the three parameters into a single number such as
# 311. The subplots will be organized in 3 rows and 1 column. Give the subplot the
# title "Polynomial". Make a solid red line
```

```python
func=np.poly1d(np.array([1,2,3,4]).astype(float))
func1=func.deriv(m=1)
x=np.linspace(-10,10,30)
y=func(x)
y1=func1(x)
func2=func.deriv(m=2)#second order derivative
y3=func2(x)
#first subplot
plt.subplot(311)
plt.plot(x,y,'g--')
plt.show()

#second subplot
plt.subplot(3,2,1)
plt.plot(x,y1,'b*')
plt.show()

#third plot
plt.subplot(3,3,1)
plt.plot(x,y3)
plt.show()

#thi
```
